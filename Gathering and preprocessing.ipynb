```python
# Importing the excel to impute the missing values in the data 
updates = pd.read_csv(r"C:\Users\surya\OneDrive\Documents\World_pop.csv")
```

```python
updates.head()
```

```python
pop = pop.merge(updates,how = 'left',on='Country')

# Handling the missing values 
pop['HI_Avg_4decades'] = pop['HI_Avg_4decades'].fillna(pop['hdi'])

pop['fertility_rate_x'] = pop['fertility_rate_x'].fillna(pop['fertility_rate_y'])

pop['birthrate_10yr_avg'] = pop['birthrate_10yr_avg'].fillna(pop['birth_rate'])

pop['gdpgrowth_10yr_avg'] = pop['gdpgrowth_10yr_avg'].fillna(pop['gdp_growth'])

pop.drop(columns=['hdi','fertility_rate_y','birth_rate','gdp_growth','death_rate'],inplace = True)
```

```python
def imputation(value):
    if value <= 0.45:
        return 'Low'
    elif 0.45 < value <= 0.58:
        return 'Medium'
    elif 0.58 < value <= 0.70:
        return 'High'
    elif value > 0.70:
        return 'Very High'
```

```python
pop['HdiTier'].fillna(pop['HI_Avg_4decades'].apply(imputation),inplace=True)
```
